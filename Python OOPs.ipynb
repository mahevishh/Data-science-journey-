## Method Arguments

```python
# Method arguments are the inputs passed to a method.
# They allow methods to receive data and perform actions based on that data.
# 'self' must always be the first parameter in instance methods.
```

```python
#Default Arguments
# Default arguments provide fallback values if no arguments are passed.
class students:
    def info(self, name='ayush', age=16):
        print(f'name:{name}, age:{age}')
s1 = students()
s1.info()
```

```python
# Positional Arguments
# Positional arguments must be passed in the same order as defined.
class mathoops:
    def sum(self, a,b):
        print('sum=', a+b)
s = mathoops()
s.sum(2,6) # Arguments matched by position
```

```python
# Keyword Arguments
# Keyword arguments specify which parameter each value corresponds to.
class mathoops:
    def sum(self, a,b):
        print(f'a is {a}, b is {b}')
s = mathoops()
s.sum(b=11, a = 9) 
```

```python
# Variable-Length Arguments (args)
# *args lets a method accept any number of positional arguments.
class cal:
    def add_all(self, *nums):
        print('total', sum(nums))
c = cal()
c.add_all(1,7,8,2,3,4)
```

```python
# Keyword Variable-Length Arguments (kwargs)
# **kwargs allows passing any number of keyword arguments as a dictionary.

class Info:
    def display(self, **data):
        for key, value in data.items():
            print(f"{key}: {value}")

i = Info()
i.display(name="alia", course="Python", level="Intermediate")
```

## Inheritance

```python
# Inheritance allows one class (child/derived class) to use the properties and methods of another class (parent/base class).
# It promotes code reusability and logical class hierarchy.
```

### 1. Single Inheritance

```python
# In Single Inheritance, one child class inherits from one parent class.
class parent:
    def p(self):
        print("i am parent class")
class child(parent):
    def c(self):
        print("i am child class")
obj = child()
obj.p()
obj.c()
```

### 2. Multiple Inheritance

```python
# In Multiple Inheritance, a child class inherits from more than one parent class.
class father:
    def method1(self):
        print("i am father class")
class mother:
    def method2(self):
        print("i am mother class")
class child(father, mother):
    def method3(self):
        print("i am child class")

c = child()
c.method1()
c.method2()
c.method3()
```

### 3. Multilevel Inheritance

```python
# In Multilevel Inheritance, a class is derived from another derived class.
class grandfather:
    def method1(self):
        print("feature from grandfather class")
class father(grandfather):
    def method2(self):
        print("feature from father class")
class child(father):
    def method3(self):
        print("feature from child class")
obj = child()
obj.method1()
obj.method2()
obj.method3()
```

### 4. Hierarchical Inheritance

```python
# In Hierarchical Inheritance, multiple child classes inherit from a single parent class.
class Parent:
    def show_parent(self):
        print("Feature from Parent")

class Child1(Parent):
    def show_child1(self):
        print("Feature from Child1")

class Child2(Parent):
    def show_child2(self):
        print("Feature from Child2")

c1 = Child1()
c2 = Child2()

c1.show_parent()
c2.show_parent()
```

 ### 5. Hybrid Inheritance

```python
# Hybrid Inheritance is a combination of two or more types of inheritance.
# It may involve complex relationships among classes.
class A:
    def showA(self):
        print("Class A")

class B(A):
    def showB(self):
        print("Class B")

class C(A):
    def showC(self):
        print("Class C")

class D(B, C):
    def showD(self):
        print("Class D")

obj = D()
obj.showA()
obj.showB()
```

## Static Methods

```python
# Static methods are defined using the @staticmethod decorator.
# They do not use 'self' or 'cls' because they don’t access instance or class data.
# They behave like normal functions but belong to a class’s namespace.
```

```python
class MathOps:
    @staticmethod
    def add(a, b):
        return a + b

print(MathOps.add(5, 7))
# Static methods can be called using the class name without creating an object.
```

```python
# Static methods help organize utility functions related to a class.
class Calculator:
    @staticmethod
    def multiply(x, y):
        return x * y

    def display_result(self):
        print("Multiplication result:", Calculator.multiply(4, 3))

c = Calculator()
c.display_result()
```

## Decorators

```python
# A decorator is a function that modifies the behavior of another function or method.
# In Python, decorators are written using the @ symbol above a function definition.
```

```python
def greet_decorator(func):
    def wrapper():
        print("Good Morning!")
        func()
        print("Have a nice day!")
    return wrapper

@greet_decorator # The @greet_decorator adds extra functionality before and after the 'hello' function.
def hello():
    print("Hello, User!")

hello()
```

```python
class Student:
    school_name = "ABC Public School"   # Class variable

    def __init__(self, name, age):
        self.name = name    # Instance variable
        self.age = age

    # Instance Method
    def show_info(self):
        print(f"Name: {self.name}, Age: {self.age}")

    # Class Method
    @classmethod
    def school(cls):
        print("School Name:", cls.school_name)

    # Static Method
    @staticmethod
    def greet():
        print("Welcome to the Student Management System!")

# Creating an object
s1 = Student("alia", 20)

# Calling instance method
s1.show_info()

# Calling class method
Student.school()

# Calling static method
Student.greet()
```

## Inner Class (Nested Class)

```python
# An inner class is a class defined inside another class.
# It helps logically group classes that are closely related and used only within the outer class.
```

```python
class Outer:
    def __init__(self):
        self.name = "Outer Class"

    def show(self):
        print("This is the outer class")

    class Inner:
        def display(self):
            print("This is the inner class")

# Creating objects
outer_obj = Outer()
outer_obj.show()

inner_obj = outer_obj.Inner()   # Accessing inner class through outer class object
inner_obj.display()
```

```python
class Company:
    def __init__(self, cname):
        self.cname = cname

    def show(self):
        print("Company Name:", self.cname)

    class Employee:
        def __init__(self, ename):
            self.ename = ename

        def display(self):
            print("Employee Name:", self.ename)

# Creating outer class object
comp = Company("TechCorp")
comp.show()

# Creating inner class object
emp = comp.Employee("ajay")
emp.display()
```

## super()

```python
# The super() function is used to call methods from a parent (base) class inside a child (derived) class.
# It helps avoid referring to the parent class name directly.
```

```python
class Parent:
    def show(self):
        print("This is the Parent class method")

class Child(Parent):
    def show(self):
        super().show()    # Call parent class method
        print("This is the Child class method")

obj = Child()
obj.show()
```

```python
#Using super() in Constructor
class Animal:
    def __init__(self, name):
        self.name = name
        print(f"Animal name is {self.name}")

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)   # Call parent constructor
        self.breed = breed
        print(f"Breed is {self.breed}")

d = Dog("Buddy", "Golden Retriever")
```

```python
#Multiple Inheritance with super()
class A:
    def display(self):
        print("Class A")

class B(A):
    def display(self):
        super().display()
        print("Class B")

class C(B):
    def display(self):
        super().display()
        print("Class C")

obj = C()
obj.display()
```

### Operator Overloading

```python
# Operator overloading lets us redefine how operators work for user-defined objects.
# This is done using special methods (dunder methods).
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

p1 = Point(2, 3)
p2 = Point(4, 1)
p3 = p1 + p2   # Calls __add__ method

print(p3.x, p3.y) # The '+' operator is overloaded to add coordinates of two Point objects.

```

##  Abstract Class & Abstract Method

```python
# An Abstract Class is a class that cannot be instantiated directly.
# It serves as a blueprint for other classes.
# Abstract Methods are methods declared in the abstract class but must be implemented in child classes.
# They are defined using the 'abc' (Abstract Base Class) module.
```

```python
from abc import ABC, abstractmethod

class Shape(ABC):       # Abstract class
    @abstractmethod
    def area(self):
        pass             # Abstract method (no implementation here)

class Circle(Shape):    # Child class
    def __init__(self, radius):
        self.radius = radius

    def area(self):      # Implementing abstract method
        return 3.14 * self.radius * self.radius


circle = Circle(5)
print("Area of circle:", circle.area())
```

```python
#Abstract Class with Multiple Abstract Methods
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass

    @abstractmethod
    def stop(self):
        pass

class Car(Vehicle):
    def start(self):
        print("Car engine started")

    def stop(self):
        print("Car engine stopped")

car = Car()
car.start()
car.stop()
```

## Iterators in Python

```python
# An iterator is an object that can be iterated (looped) upon.
# It implements two methods: __iter__() and __next__().
# Iterators return one element at a time when next() is called.
```

```python
my_list = [1, 2, 3]
it = iter(my_list)       # Create iterator object

print(next(it))          # Output: 1
print(next(it))          # Output: 2
print(next(it))          # Output: 3
# print(next(it))        # Raises StopIteration
```

```python
class Counter:
    def __init__(self, start, end):
        self.current = start
        self.end = end

    def __iter__(self):
        return self

    def __next__(self):
        if self.current <= self.end:
            num = self.current
            self.current += 1
            return num
        else:
            raise StopIteration

for num in Counter(1, 5):
    print(num)
```

## Generators in Python

```python
# A generator is a simpler way to create an iterator using the 'yield' keyword.
# It automatically creates __iter__() and __next__() methods.
# Generators are memory efficient since they yield one value at a time.
```

```python
def simple_gen():
    for i in range(1, 4):
        yield i   # yield pauses the function and returns a value

gen = simple_gen()
print(next(gen))
print(next(gen))
print(next(gen))
# print(next(gen))  # Raises StopIteration
# Each call to next() resumes the function from where it left off.
```

```python
def squares(n):
    for i in range(1, n+1):
        yield i*i

for val in squares(5):
    print(val)
```

## Multithreading

```python
# Multithreading allows multiple parts (threads) of a program to run concurrently.
# It helps perform multiple tasks at the same time — improving performance in I/O-bound programs.
```

```python
import threading

def num():
    for i in range(1,6):
        print(i)
def char():
    for j in ['a','b','c','d']:
        print(j)
t1 = threading.Thread(target=num)
t2 = threading.Thread(target=char)
# Starting threads
t1.start()
t2.start()

# Waiting for both threads to complete
t1.join()
t2.join()

print("Both threads completed!")
```

```python
import threading
import time

def worker():
    print("Thread started")
    time.sleep(2)
    print("Thread ended")

thread = threading.Thread(target=worker)
thread.start()

print("Active threads:", threading.active_count())
```
